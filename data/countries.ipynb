## Regions & Comparisons

```python
REGIONS = {
    "Africa": ["ZAF", "NGA"],
    "Asia": ["CHN", "JPN", "IND", "IDN", "THA", "VNM", "MYS", "PHL", "PAK", "BGD"],
    "Europe": ["DEU", "GBR", "FRA", "ITA", "ESP", "POL", "RUS", "TUR"],
    "North America": ["USA", "CAN", "MEX"],
    "South America": ["BRA", "ARG"],
    "Middle East": ["SAU", "ARE", "EGY"],
    "Oceania": ["AUS"],
}

DEFAULT_COMPARISONS = {
    "EGY": ["IND", "SAU", "ARE", "WLD"],
    "IND": ["EGY", "CHN", "USA", "WLD"],
    "SAU": ["EGY", "ARE", "IRQ", "WLD"],
}
```

## Helper Functions

```python
def get_country_region(code):
    return COUNTRIES.get(code.upper(), {}).get("region", "Unknown")


def get_all_regions():
    return list(REGIONS.keys())


def get_countries_in_region(region):
    return REGIONS.get(region, [])


def get_default_comparisons(code):
    return DEFAULT_COMPARISONS.get(code.upper(), ["WLD"])
```

## Test & Explore

```python
print(f"Total countries: {len(COUNTRIES)}")
print(f"\nAll country codes:")
for code in sorted(COUNTRIES.keys()):
    print(f"  {code} — {COUNTRIES[code]['region']}")
```

```python
print("Regions and their countries:")
for region, codes in REGIONS.items():
    print(f"\n{region}:")
    print(f"  {', '.join(codes)}")
```

```python
print("Default comparisons:")
for code, comps in DEFAULT_COMPARISONS.items():
    print(f"  {code}: {comps}")
```
