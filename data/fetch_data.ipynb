# Fetch OWID CO₂ Emissions Data

Downloads CO₂ emissions data from [Our World in Data](https://github.com/owid/co2-data) and saves it to `data/cache/owid_emissions.csv`.

**Source:** `https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv`

**Note:** Re-run this notebook whenever you want fresh data. It uses a cache — if the file already exists, it loads from there instead of re-downloading.

```python
import pandas as pd
import os

OWID_URL = "https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv"
SAVE_PATH = "data/cache/owid_emissions.csv"

os.makedirs(os.path.dirname(SAVE_PATH), exist_ok=True)

if os.path.exists(SAVE_PATH):
    print(f"Cache found at {SAVE_PATH} — skipping download")
else:
    print("Fetching OWID data from GitHub...")
```

## Fetch & Save

Downloads the CSV from GitHub if not already cached, then saves it locally.

```python
if not os.path.exists(SAVE_PATH):
    df = pd.read_csv(OWID_URL)
    df.to_csv(SAVE_PATH, index=False)
    print(f"Saved {len(df)} rows to {SAVE_PATH}")
else:
    df = pd.read_csv(SAVE_PATH)
    print(f"Loaded from cache: {len(df)} rows")
```
